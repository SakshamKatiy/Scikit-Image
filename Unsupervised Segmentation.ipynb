```python
from skimage.segmentation import slic,mark_boundaries
from skimage import data
import matplotlib.pyplot as plt
from skimage.color import label2rgb
```

```python
astro = data.astronaut()
plt.figure(figsize=(15,15))
astro_segments = slic(astro,n_segments=100,compactness=1)
plt.subplot(1,2,1)
plt.imshow(astro)
plt.subplot(1,2,2)
plt.imshow(mark_boundaries(astro,astro_segments))
```

```python
astro = data.astronaut()
plt.figure(figsize=(15,15))
astro_segments = slic(astro,n_segments=50,compactness=10)
plt.subplot(1,2,1)
plt.imshow(astro)
plt.subplot(1,2,2)
plt.imshow(label2rgb(astro_segments,astro,kind = 'avg'))
```
