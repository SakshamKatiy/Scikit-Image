```python
from skimage.segmentation import felzenszwalb,mark_boundaries
from skimage.color import label2rgb
from skimage.data import astronaut
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
astro = astronaut()
astro_segment = felzenszwalb(astro,scale = 2,sigma =5,min_size = 200)
plt.subplot(1,2,1)
plt.imshow(astro)
plt.subplot(1,2,2)
plt.imshow(astro)
plt.subplot(1,2,2)
plt.imshow(mark_boundaries(astro,astro_segment))
```
